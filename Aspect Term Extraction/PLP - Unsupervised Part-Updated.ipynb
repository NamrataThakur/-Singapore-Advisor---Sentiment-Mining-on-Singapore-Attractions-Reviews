```python
import torch
from transformers import BertTokenizer

# Load the BERT tokenizer.
print('Loading BERT tokenizer...')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
def encode(docs):
    '''
    This function takes list of texts and returns input_ids and attention_mask of texts
    '''
    encoded_dict = tokenizer.batch_encode_plus(docs, add_special_tokens=True, max_length=512, padding='max_length',
                            return_attention_mask=True, truncation=True, return_tensors='pt')
    input_ids = encoded_dict['input_ids']
    return input_ids
```

```python
data_embeddings= encode(data_texts)
print(len(data_embeddings))
```

```python
#n_neighbors=15, n_components=5,
umap_embeddings = umap.UMAP(n_neighbors=30, 
                            n_components=10, 
                            metric='cosine').fit_transform(data_embeddings)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.cluster import KMeans
```

```python
#Creating the StandardScaler 
scaler = StandardScaler(with_mean=False)

#Fitting the scaler on the dataset
scaler.fit(umap_embeddings)

#Transforming the corpus with the appropiate mean and standard deviation (as found in standard scaler)
kMeans_umap_scaled = scaler.transform(umap_embeddings)

#Checking the shape. It should remain same.
print(kMeans_umap_scaled.shape)
```

```python
#Selecting hyper-parameter value
clusters = 10

#Creating a list to contain the inertia values for each hyper-parameter
train_inertia = []

#Creating a list to contain the silhouette scores for each hyper-parameter
train_silhouette_score = []

#Iterating for each value of hyper-parameter
for i in tqdm(range(2,clusters+1)):
    
    #Creating the KMeans object for each cluster
    km = KMeans(n_clusters=i, random_state=0)
    
    #Fitting the KMeans on the BoW Scaled Data and getting the cluster labels
    labels = km.fit_predict(kMeans_umap_scaled)
    
    #Getting the silhouette score for the labels
    silhouette_score = metrics.silhouette_score(kMeans_umap_scaled, labels)
    
    #Printing the Inertia values for each cluster
    print("For n_clusters =",i,"The Inertia Score is :",km.inertia_)  
    
    #Appending the inertia values to the list
    train_inertia.append(km.inertia_)
    
    #Printing the silhouette score for each cluster
    print("For n_clusters =",i,"The average silhouette_score is :",silhouette_score)
    
    #Appending the silhouette score to the list
    train_silhouette_score.append(silhouette_score)
```

```python
# Plotting Inertia_ against Number of CLusters
plt.plot(range(2,clusters+1),train_inertia)
plt.title("INERTIA PLOTS for Umap Embeddings")
plt.xlabel('Hyper-parameter Cluster Number')
plt.ylabel('Inertia Value')
plt.show()
```

```python
# Plotting Silhouette Score against Number of CLusters
plt.plot(range(2,clusters+1),train_silhouette_score)
plt.title("Silhouette Score Plot for Umap Embeddings")
plt.xlabel('Hyper-parameter Cluster Number')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
def deContract(review):
    review = re.sub("\S*\d\S*", "", review).strip()
    review = re.sub(r'[^a-zA-Z]+', ' ',review)
    review = re.sub(r"doesn\'t", "does not",review)
    review = re.sub(r"aren\'t", "are not", review)
    review = re.sub(r"it\'s", "it is", review)
    review = re.sub(r"won\'t", "will not", review)
    review = re.sub(r"can\'t", "can not", review)
    review = re.sub(r"n\'t", " not", review)
    review = re.sub(r"\'re", " are", review)
    review = re.sub(r"it\'s", " it is", review)
    review = re.sub(r"\'d", " would", review)
    review = re.sub(r"\'ll", " will", review)
    review = re.sub(r"\'ve", " have", review)
    review = re.sub(r"\'m", " am", review)
    review = re.sub(r"\'s", "s", review)
    review = re.sub(r"wouldn\'t", "would not", review)
    review = re.sub(r"isn\'t", "is not", review)
    review = re.sub(r"haven\'t", "have not", review)
    review = re.sub(r"shan\'t", "shall not", review)
    review = re.sub(r"couldn\'t", "could not", review)
    review = re.sub(r"hadn\'t", "had not", review)
    review = re.sub(r"weren\'t", "were not", review)
    review = re.sub(r"don\'t", "do not", review)
    review = re.sub(r"didn\'t", "did not", review)
    review = re.sub(r"mustn\'t", "must not", review)
    review = re.sub(r"hasn\'t", "has not", review)
    review = re.sub(r"wasn\'t", "was not", review)
    return review
```

```python
def data_preprocessing(text):
    clean_text = []
    for review in tqdm(text):
        review = deContract(review)
        stopwords_en_withpunct = set(stopwords).union(set(punctuation))
        #Remove numbers,whitespace,special characters (anything other than A-Za-z)
        review = re.sub("\S*\d\S*", "", review).strip()
        review = re.sub('[^A-Za-z]+', ' ', review)
        
        #For every word in the sentence check if is a Stopword and it's length less than or equal to 2, 
        #if not convert it to lower case
        review = ' '.join(e.lower() for e in review.split() if e.lower() not in stopwords_en_withpunct and len(e.lower()) > 3)
        clean_text.append(review)
    return clean_text
```

```python
wnl = nltk.WordNetLemmatizer()
def pos_tagger(review):
    lemma_text = []
    #print(review)
    text = nltk.word_tokenize(review)
    pos_text = nltk.pos_tag(text)
    for word, tag in pos_text:
        if tag[0] in ['N']:
            if tag in ('NN','NNPS','NNP','NNS'):
                lemma_text.append(wnl.lemmatize(word,pos= tag[0].lower()))
            else:
                lemma_text.append(word)
    return lemma_text
```

```python
def lemmatization_tagger(text):
    lemma_text = []
    #print(type(text))
    for review in tqdm(text):
        clean_text = pos_tagger(review)
        lemma_text.append((' '.join(clean_text)))
    return lemma_text
```

```python
def pre_process(text):
    tokens = nltk.word_tokenize(text)
    tokens = [ t for t in tokens if len(t) >= 3 ]
    return(tokens)
```

CLUSTER 0

Tuning the Gensim's TM to get best topic numbers

```python
# Use Gensim's NMF to get the best num of topics via coherence score
text = df_kMeansCluster0['final_clean_review_text']
texts = text.apply(pre_process)
dictionary = corpora.Dictionary(texts)
dictionary.filter_extremes(no_below=3,no_above=0.85,keep_n=5000)
corpus = [dictionary.doc2bow(text) for text in texts]
# List of the topic numbers
topic_nums = list(np.arange(5, 55 + 1, 5))

# Run the nmf model and calculate the coherence score
# for each number of topics
coherence_scores = []

for num in topic_nums:
    nmf = Nmf(corpus=corpus,num_topics=num,id2word=dictionary,chunksize=2000,
        passes=5,
        kappa=.1,
        minimum_probability=0.01,
        w_max_iter=300,
        w_stop_condition=0.0001,
        h_max_iter=100,
        h_stop_condition=0.001,
        eval_every=10,
        normalize=True,
        random_state=42
    )
    
    # Run the coherence model to get the score
    cm = CoherenceModel(
        model=nmf,
        texts=texts,
        dictionary=dictionary,
        coherence='c_v'
    )
    
    coherence_scores.append(round(cm.get_coherence(), 5))

scores = list(zip(topic_nums, coherence_scores))
best_num_topics = sorted(scores, key=operator.itemgetter(1), reverse=True)[0][0]

print(best_num_topics)
```

```python
def pre_process(text):
    #print(text)
    tokens = nltk.word_tokenize(text)
    tokens = [ t for t in tokens if len(t) >= 3 or t is not np.isnan]
    return(tokens)
```

```python
print(df_kMeansCluster1['final_clean_review_text'].isnull().sum())
df_kMeansCluster1.dropna(subset = ['final_clean_review_text'], inplace=True)
print(df_kMeansCluster1['final_clean_review_text'].isnull().sum())
```

```python
# Use Gensim's NMF to get the best num of topics via coherence score
text = df_kMeansCluster1['final_clean_review_text']
texts = text.apply(pre_process)
dictionary = corpora.Dictionary(texts)
dictionary.filter_extremes(no_below=3,no_above=0.85,keep_n=5000)
corpus = [dictionary.doc2bow(text) for text in texts]
# Create a list of the topic numbers we want to try
topic_nums = list(np.arange(5, 55 + 1, 5))

# Run the nmf model and calculate the coherence score
# for each number of topics
coherence_scores = []

for num in topic_nums:
    nmf = Nmf(corpus=corpus,num_topics=num,id2word=dictionary,chunksize=2000,
        passes=5,
        kappa=.1,
        minimum_probability=0.01,
        w_max_iter=300,
        w_stop_condition=0.0001,
        h_max_iter=100,
        h_stop_condition=0.001,
        eval_every=10,
        normalize=True,
        random_state=42
    )
    
    # Run the coherence model to get the score
    cm = CoherenceModel(
        model=nmf,
        texts=texts,
        dictionary=dictionary,
        coherence='c_v'
    )
    
    coherence_scores.append(round(cm.get_coherence(), 5))

scores = list(zip(topic_nums, coherence_scores))
best_num_topics = sorted(scores, key=operator.itemgetter(1), reverse=True)[0][0]

print(best_num_topics)
```
